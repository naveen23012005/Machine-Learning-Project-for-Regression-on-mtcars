#### Removing Duplicates

```python
mtcars=mtcars.drop_duplicates()
```

#### Split the data into x,y

```python
x=mtcars.iloc[:,1:]
y=mtcars.iloc[:,0]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
x_train.head()
```

```python
x_test.head()
```

#### Identify Numerical and Categorical columns in the Data

```python
numerical = x_train.select_dtypes(include=['int64','float64']).columns
numerical
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
linear_pred = linear_model.predict(x_test)
```

```python
random_model = RandomForestRegressor()
random_model.fit(x_train, y_train)
random_pred = random_model.predict(x_test)
```

```python
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    print("MSE:", mse)
    print("RMSE:", rmse)
    print("R2 Score:", r2)
```

```python
# "Linear Regression"
```

```python
evaluate_model(y_test,linear_pred)
```

```python
#"Random Forest"
```

```python
evaluate_model(y_test,random_pred)# based on cross validation random forest better than linear regression 
```
